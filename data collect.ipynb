# Cleaning the Dataset

```python
# Dropping these rows because either their values are NaN or not useful for us

df = df.drop(["sid","id" , "position","created_at","created_meta","updated_at","updated_meta","meta"],axis=1)
```

```python
#Noticed that a lot of values got corrupted and got printed as "s" 
#So replace "s"  and "No Data" with NaN
df=df.replace("s",np.nan)
df=df.replace("No Data",np.nan)
```

```python
df
```

```python
#There are columns which shows the percentage of their respective columns . Example :" Asian " and " %Asian "
#We dont need these columns as we are getting the required information from their count columns

#We won't be removing %#Male & %#Female now as we require them later 

df= df.drop(["%Asian","%Black","%Hispanic","%Other","%White ","%ELL Bengali","%ELL Spanish"
         ,"%ELL Chinese","%ELL Arabic","%ELL Haitian Creole","%ELL French","%ELL Russian","%ELL Korean","%ELL Urdu",
        "%ELL Other","ELA %Level 1","ELA %Level 2","ELA %Level 3","ELA %Level 4","ELA %L3+L4",
        "MATH %Level 1","MATH %Level 2","MATH %Level 3","MATH %Level 4","MATH %L3 + L4"],axis=1)
```

```python
#Check if there's any duplicate values
duplicate = df.duplicated()
df[duplicate]
```

```python
#Missing Values
total = df.isnull().sum().sort_values(ascending = True)
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending = True)
missing_tab = pd.concat([total,percent],axis=1,keys=["Total","Percent"])
print(missing_tab.head(50))
```

```python
#Lets drop columns with very high null values
df = df.drop(["#Other","#ELL Spanish ","#ELL Other","#ELL Arabic","#ELL Chinese","#ELL Bengali","#ELL Urdu","#ELL Russian",
              "#ELL Haitian Creole","#ELL Korean","#ELL French","ELA #Level 4","ELA #Level 3","MATH #Level 3",
             "MATH #Level 4","#Grade 8","#Grade 7","#Grade 6","#Asian","#White","#Grade 5","#Grade 4",
             "#Grade K","#Grade 3","#Grade 2","#Grade 1","ELA #L3+L4","MATH #L3+L4"],axis=1)
```

Total Enrollment

```python
#["Total Enrollment"] is an important column because it tells us how many students are enrolled from each school
#Now , lets check the row values for other columns when the ["Total Enrollment"] is a null value

df[df["Total Enrollment"].isnull()].info()
```

```python
#So when ["Total Enrollment"] is null , majority of other columns are null too.
#We cant do any7thing with the data when the no of enrolled students are not given
#So , lets consider only the rows where "Total Enrollment" is not null
df = df[df["Total Enrollment"].notna()]
```

Male and Female

```python
#Lets check next columns
df[df["#Male"].isnull()].info()
```

```python
#We can see equal no.of null values for male and female columns
#Also notice that for these rows Total Enrollment values are provided.
#So we can divide Total Enrollment values to Male and Female columns respectively.
#To do that , first we need to convert datatype of the required columns from object to float/int
df["Total Enrollment"] = df["Total Enrollment"].astype("int") 
df["#Male"] = df["#Male"].astype("float") 
df["#Female"] = df["#Female"].astype("float") 

df["%Male"] = pd.to_numeric(df["%Male"], errors='coerce')
df["%Female"] = pd.to_numeric(df["%Female"], errors='coerce')
```

```python
#Lets check the average Female and Male Percentage
```

```python
df["%Female"].mean()
```

```python
df["%Male"].mean()
```

```python
#Now we have to divide the Total enrollment values such that 45% is Female and 55% is Male
#Lets check how many rows contains Null values for #Male and #Female columns
cols = ['#Male', '#Female',]
df[df[cols].isnull().any(axis=1)]
```

```python
v = df['Total Enrollment']
df[["#Female"]] = df[["#Female"]].mask(df.isna(),(5*v/11),axis=0)
```

```python
df["#Female"] = df["#Female"].astype("int")
#Converting #Female values to integer so that while subtracting in the next step we get answer as an integer.Also no.of person cannot be in decimal.
```

```python
x= df["#Female"]
df[["#Male"]] = df[["#Male"]].mask(df.isna(),(v-x),axis=0)
df["#Male"] = df["#Male"].astype("int")
df[["Total Enrollment",'#Male','#Female']].head(40)
```

```python
#Since we have fixed the #Male and #Female columns , we dont need their percentage columns anymore
df = df.drop(["%Female","%Male"],axis=1)
```
